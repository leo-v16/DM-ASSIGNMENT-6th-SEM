```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('spam.csv')
df.describe(), df.head()
```

```python
df['spam'] = df['Category'].map({'ham': 0, 'spam': 1})
df.head()
```

```python
cv = CountVectorizer()
X = cv.fit_transform(df['Message'])
y = df['spam']
```

```python
model = MultinomialNB()
model.fit(X, y)
model.score(X, y)
```

```python
message = df['Message'][0]
message_vector = cv.transform([message])
prediction = model.predict(message_vector)
print(f'Message: {message}')
print(f'Predicted as: {"Spam" if prediction[0] == 1 else "Ham"}')
```
